```r
library(readr)
library(tidyverse)
library(VIM)
library(forcats)
setwd("C:/Users/iceca/Documents/Housing_Regression")
load("Data/RDAs/rawTrain.rda")
load("Data/RDAs/rawTest.rda")
getwd()
```

```r
raw.train %>% ncol()
raw.train %>% nrow()
raw.test %>% ncol()
raw.test %>% nrow()

#In order to make the preprocessing easier I am going to combine the training set and text set temporarily, 
    #to do this I will create a column for isTrain
raw.train$isTrain = 1
raw.test$isTrain = 0

labels <- data.frame(SalePrice = raw.train$SalePrice , Id=raw.train$Id) 

ids.test <- raw.test$Id
save(ids.test, file = "Data/RDAs/IdTest.rda")


raw.train$SalePrice <- NULL


raw <- full_join(raw.train, raw.test)
#Only numeric and character vectors are in the raw matrix
classes <- sapply(raw.train, class)

cols.num <- names(raw)[classes == 'numeric']
nums <- raw %>% select(cols.num)

cols.char <- names(raw)[classes != 'numeric']
characters <- raw %>% select(cols.char)
characters$isTrain <- nums$isTrain
```

```r
#I was originally planning delete the Ids column, but instead through out this notebook I arrange the data by Id. I do this
#to ensure that the instances are not mixed up when using cbind. 
```

```r
#REPLACING NA IN NUMS
#apply(nums, 2, function(x) {sum(is.na(x))})


# GarageYearBuilt and LotFrontage have over 100 missing values, should we remove them
# LotFrontage --> this is probably not super important, but I think it is predictable given that newer home have less lots size
# GarageYearBuilt --> we will regress its value, I think it is easy to predict
# sum( nums$GarageYrBlt == nums$YearBuilt , na.rm=TRUE) #For over 2000 values the garage year built is same as house built
# sum( nums$GarageYrBlt == nums$YearRemodAdd , na.rm=TRUE) #For over 100 values the garageyear built is same as remodel year

# GarageYrBlt highly correlated with yearbuilt, as hypothesised
gybCols <- names(nums)[apply (nums, 2, function(x) {cor( nums$GarageYrBlt , x,  use="complete.obs" )}) >0.55] 
# hence i will fit a linear regression with the most correlated variables
gybDat <- nums %>% filter(isTrain==1) %>% select(gybCols)
gybFit <- lm(GarageYrBlt ~ ., data= gybDat)
raw.missingGYB <- nums %>% filter(is.na(GarageYrBlt))
nums <- nums %>% filter(!is.na(GarageYrBlt))
raw.missingGYB$GarageYrBlt <- predict(gybFit, raw.missingGYB)
nums <- rbind(nums, raw.missingGYB)
# Lot Frontage is not correlated with any variables, so it will be hard to predict, my hypothesis was wrong.
names(nums)[apply (nums, 2, function(x) {cor( nums$LotFrontage , x,  use="complete.obs" )}) >0.5]
#i think it would be best to replace with the average

#now replace any NA's with the value of the mean
num.means <- apply(nums %>% filter(isTrain==1), 2, function(x) {mean(x,na.rm=TRUE)})
for(i in 1:ncol(nums)){
  nums[is.na(nums[,i]), i] <- num.means[i]
}
nums <- nums %>% arrange(Id)
```

```r
#REPLACING NA'S IN factors FEATURES

factors <- mutate_all(characters, factor)
cat("There are only", nrow(factors), "rows")
how.many.nas <- apply(factors 
      , 2, function(x) {sum(is.na(x))})
levs <- rep(NA , ncol(factors))
for (i in 1:ncol(factors)) {
    levs[i] <- length(levels(factors[,i] %>% pull(1)))    
}
summary.factors <- data.frame(numLevels = levs , numNA = how.many.nas)
summary.factors

#remove the features with over 1000 na's 
factors <- factors %>% select_if(function(x) {
    sum(is.na(x )) <  1000          }) %>% mutate(isTrain=nums$isTrain, Id=nums$Id, MiscFeature=factors$MiscFeature)

#I was originally looking to remove the rows with nas in columns that have few nas in total,
#however I realized the most of those NA are from the test set, so i left the NA
dftr <- data.frame(trainn = apply( factors %>% filter(isTrain==1), 2, function(x) {sum(is.na(x))}) , 
          testt = apply( factors %>% filter(isTrain!=1), 2, function(x) {sum(is.na(x))})
          )  


#for all the other columns make a new na factor because there arent too many of them but they are still
#significant. furthermore if you read the data description na are often indicate not applicable, for example
#na for pool quality indicates no pool.
nas <- factors %>% select_if(function(x) {
    sum(is.na(x )) >0      })
nas <- mutate_all(nas, as.character)
nas <- nas %>% replace(., is.na(.), "Unknown")
nas <- mutate_all(nas, factor)
nas$Id <- factors$Id
factors <- factors[, !(names(factors) %in% names(nas)) | (names(factors) %in% c("Id")) ]
nas <- nas %>% arrange(Id) %>% select(-Id)
factors <- factors %>% arrange(Id)
factors <- cbind(factors , nas)
```

```r
#some features have too many levels
too.many.factors <- factors %>% select_if(function(x) {length(levels(x))>=7})%>% 
    mutate(isTrain = nums$isTrain, Id= nums$Id)
sapply(too.many.factors, function(x) {length(levels(x))})
too.many.factors


#EXTERIOR
dat <- too.many.factors %>% filter(isTrain==1) %>% mutate(labels = labels) %>% group_by(Exterior1st) %>%
    summarize(prop = mean(labels))
ggplot(data=dat, aes(x=Exterior1st, y=prop, fill=Exterior1st)) + geom_bar(stat="identity") + 
    theme(axis.text.x=element_text(angle=90, hjust=1)) +
    geom_abline(slope=0, intercept=2*10^5,  col = "black",lty=5)+
    geom_abline(slope=0, intercept=1.2*10^5,  col = "black",lty=5)+
    geom_text(aes(label=Exterior1st, angle=90))
#seems like there are 3 groups
low=c("AsbShng" , "BrkComm", "AsphShn", "CBlock")
high=c("CemntBd" , "ImStucc", "Stone", "VinylSd") #and everything else is medium
too.many.factors <- too.many.factors %>% mutate(Exterior1st = ifelse(Exterior1st %in% low, "low",
            ifelse(Exterior1st %in% high, "high", "medium") ) , 
              Exterior2nd = ifelse(Exterior2nd %in% low, "low",
            ifelse(Exterior2nd %in% high, "high", "medium") )                        
            )

#BASEMENT FINISH TYPE
(too.many.factors %>% filter(isTrain==1)) %>% mutate(labels=labels$SalePrice) %>% 
    group_by(BsmtFinType1) %>%  summarize(prop=mean(labels))
#I will group into no basement, good, average, low, unfinished and none
too.many.factors <- too.many.factors %>% mutate(BsmtFinType1 = ifelse(BsmtFinType1 %in% c("GLQ"), "good",
            ifelse(BsmtFinType1 %in% c("ALQ", "Rec"), "average",
            ifelse(BsmtFinType1 %in% c("LwQ", "BLQ"), "low", BsmtFinType1))) , 
                                                                     
            BsmtFinType1 = ifelse(BsmtFinType2 %in% c("GLQ"), "good",
            ifelse(BsmtFinType2 %in% c("ALQ", "Rec"), "average",
            ifelse(BsmtFinType2 %in% c("LwQ", "BLQ"), "low", BsmtFinType2)))                    
            )

#Functional
too.many.factors$Functional <- fct_collapse(too.many.factors$Functional ,minor=c("Min1",
                                                                                "Min2"),
                                           moderate=c("Mod"), major=c("Maj1","Maj2"), 
                                           Sev=c("Sev"), Sal=c("Sal"))

#GarageType, I don't think it is very useful
too.many.factors$GarageType <-NULL

#Roof Matl
#table(too.many.factors$RoofMatl) #basically only one main CompShg, everything else is rare
too.many.factors$RoofMatl <- fct_collapse(too.many.factors$RoofMatl ,Other=c("ClyTile","Membran", 
                                                                                "Metal", "Roll", "Tar&Grv", 
                                                                                "WdShake", "WdShngl"))

factors <- factors[, !(names(factors) %in% names(too.many.factors)) ]
factors <- cbind(factors , too.many.factors %>% arrange(Id))
factors <- mutate_all(factors, factor)
```

```r
#other factor variables that I believe are more of checkmark features, the quality is not imporatant but rather whether
#everything "works"


#I know that when buying homes GarageCond/GarageQuality is a checkmark feature, hence i will group into normal, poor, and none
factors$GarageCond <- fct_collapse(factors$GarageCond ,fine=c("Ex", "Gd","Fa", "TA"),
                                            bad=c("Po"), Unknown="Unknown")
factors$GarageQual <- fct_collapse(factors$GarageQual ,fine=c("Ex", "Gd","Fa", "TA"),
                                            bad=c("Po"), Unknown="Unknown")
levels(factors$GarageQual)
#I know that when buying homes HeatingQC is not super important, hence I will group into good, and bad
levels(factors$HeatingQC)
factors$HeatingQC <- fct_collapse(factors$HeatingQC ,fine=c("Ex", "Gd","Fa", "TA"),
                                            bad=c("Po"), Unknown="Unknown")
levels(factors$HeatingQC)

#Electrical, not important
factors$Electrical <-NULL

#Foundation, Not important
factors$Foundation <- NULL

#BsmtQual, the height of the basement can be grouped into 4 groups
factors$BsmtQual <- fct_collapse(factors$BsmtQual ,great=c("Ex", "Gd"),
                                 normal=c("Fa", "TA"),
                                            bad=c("Po"), Unknown="Unknown")
levels(factors$BsmtQual)

factors <- mutate_all(factors, factor)
```

```r
factors$isTrain <- NULL #the nums data frame already has isTrain
nums <- nums %>% mutate_all(as.numeric)
raw <- cbind(nums %>% arrange(Id), factors %>% arrange(Id) %>% select(-Id))
save(raw, file = "Data/RDAs/raw.rda")
train <- raw %>% filter(isTrain == 1)
save(train, file = "Data/RDAs/Train.rda")
test <- raw %>% filter(isTrain == 0)
save(test, file = "Data/RDAs/Test.rda")
save(labels, file = "Data/RDAs/Labels.rda")
```
